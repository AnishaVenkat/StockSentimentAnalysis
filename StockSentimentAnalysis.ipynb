# In this project we shall the stock sentiment analysis classification problem:

#There are 2 labels 0 and 1 '0' indicates stock prices going up and 1 indicates stock prices going down.

```python
import pandas as pd
```

```python
data=pd.read_csv("Data.csv",encoding="ISO-8859-1")
```

```python
data.head()
```

```python
data.shape
```

```python
data.head()
```

```python
# Removing extra spaces and punctuations:

data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)
```

```python
data['Date']
```

```python
# Splitting the training and test data:

train = data[data['Date'] < '2015-01-01']
test = data[data['Date'] > '2014-12-31']
```

```python
train.head()
```

```python
train.shape
```

```python
test.head()
```

```python
test.shape
```

```python
#Splitting independant and dependant features:

Xtrain=train.iloc[:,2:27]
Ytrain=train['Label']

Xtest=test.iloc[:,2:27]
Ytest=test['Label']
```

```python
Xtrain.head(2)
```

```python
#Merging all the columns into one and converting into lower case:

headlines=[]
for i in range(len(Xtrain.index)):
    headlines.append(' '.join(str(x) for x in Xtrain.iloc[i,:len(Xtrain.columns)]))
```

```python
for i in range(len(headlines)):
    headlines[i]=headlines[i].lower()
```

```python
headlines[0]
```

```python
len(headlines)
```

```python
# Now converting the same for test data-set

headlines_test=[]
for i in range(len(Xtest.index)):
    headlines_test.append(' '.join(str(x) for x in Xtest.iloc[i,:len(Xtest.columns)]))
for i in range(len(headlines_test)):
    headlines_test[i]=headlines_test[i].lower()
```

```python
headlines_test[0]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,20))

sns.countplot(data['Label'])
```

```python
#Inference :The target feature is perfectly balanced in this dataset.
```

```python
#Now lets apply NLP Count vectorizer for identifying relations betweens words for ngram=(3,3)

from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(ngram_range=(3,3))

train_dt=cv.fit_transform(headlines)

test_dt=cv.transform(headlines_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier(n_estimators=100,criterion='entropy')

rfc.fit(train_dt,Ytrain)

predictions=rfc.predict(test_dt)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

accuracy=accuracy_score(Ytest,predictions)
print(accuracy)
print("*"*80)

classification_report=classification_report(Ytest,predictions)
print(classification_report)
print("*"*80)

confusion_matrix=confusion_matrix(Ytest,predictions)
print(confusion_matrix)
print("*"*80)

```
